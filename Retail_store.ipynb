```python
# Import libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
data = pd.read_csv('retail.csv',encoding='unicode_escape')
data.head()
```

### Data Cleaning Processes


```python
#Check Country Column
data['Country'].value_counts()
```

##### Change Countries to full names e.g UK to United Kingdom, USA to United States of America

```python
data['Country'] = data['Country'].replace({'UK':'United Kingdom','EIRE':'Ireland','USA':'United States of America','RSA':'South Africa'})
```

```python
# Check
data['Country'].value_counts()
```

```python
# Check Description
data['Description'].value_counts()
```

```python
# Change all to Caps
data['Description'] = data['Description'].str.upper()
```

```python
# check
data['Description']
```

```python
#check Customer ID
data['Customer ID']
```

```python
# To remove the full stop and last 0
# First Convert to String 
data['Customer ID'] = data['Customer ID'].astype('str')
```

```python
# Replace all full stop with nothing
data['Customer ID'] = data['Customer ID'].str.replace('.','')
```

```python
#remove last zero
data['Customer ID'] = data['Customer ID'].str[:5]
```

```python
#check
data['Customer ID']
```

#### Convert InvoiceDate Column to DateTime DataType

```python
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])
```

```python
#Check
data.info()
```

#### To remove last alphabet from the invoice number

```python
data['StockCode']
```

```python
data['StockCode'] = data['StockCode'].str[:5]
```

##### Total amount on each stock is quantity * price

```python
data['Total Amount'] = data['Quantity'] * data['Price']
```

```python
data['Total Amount']
```

##### Total amount by Invoice 


```python
Amount_by_Invoice = data.groupby(by='Invoice')[['Total Amount']].sum()
```

```python
Amount_by_Invoice
```

##### Quantity and Total Sales per Country

```python
QT_per_Country = data.groupby(by='Country')[['Total Amount','Quantity']].sum()
```

```python
QT_per_Country
```

```python
QT_per_Country_sorted = QT_per_Country.sort_values(by='Quantity', ascending=False)
```

```python
QT_per_Country_sorted
```
